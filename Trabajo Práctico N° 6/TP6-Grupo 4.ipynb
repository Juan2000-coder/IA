<h2> Ejercicios Teóricos </h2>

1. ¿En qué tipo de algoritmos se basa un planificador para encontrar el mejor camino a un estado solución?

  Un planificador se basa en algoritmos de busqueda debido a que tenemos estados iniciales y finales, y debemos encontrar el mejor camino para llegar al estado final.


2. ¿Qué tres elementos se encuentran dentro de una acción formulada en lenguaje STRIPS? Describa brevemente qué función cumple cada uno.

  **Acción:** Este elemento identifica la acción que se va a realizar. Es un nombre que describe la acción de manera concisa, como "Mover", "Dejar", etc.

  **Precondiciones:** Las precondiciones son las condiciones que deben ser verdaderas antes de que la acción pueda ejecutarse. En otras palabras, son las condiciones que deben cumplirse para que la acción tenga sentido y pueda llevarse a cabo de manera efectiva. Por ejemplo, si la acción es "Mover", una precondición podría ser "El objeto a mover está en la ubicación actual".

  **Efectos:** Los efectos son las condiciones que serán verdaderas después de que la acción se haya ejecutado con éxito. Estas condiciones representan el cambio en el estado del mundo causado por la acción. Por ejemplo, si la acción es "Mover", un efecto podría ser "El objeto ha sido movido a una nueva ubicación".


3. Describa las ventajas y desventajas de desarrollar un algoritmo de planificación hacia adelante y hacia atrás en el espacio de estados.

  La desventaja de la planificación atras es que es muy complejo de pensar, perlo tiene la ventaja es que evaluamos menos nodos.


4. ¿Qué heurística debería emplear un planificador hacia adelante para garantizar encontrar el mejor resultado posible?

7. Suponga que tiene un robot de oficina capaz de moverse y tomar y depositar objetos. El robot solo puede tener un objeto a la vez, pero puede conseguir una *caja* en la que depositar varios objetos. Suponga que programa al robot para *ir a la tienda* a comprarle un *café* y en el camino de vuelta tome una *carta* del *buzón* de la oficina para que se la traiga junto con el café. Describa en lenguaje STRIPS:

        7.1 El dominio del robot (nombre, predicados y acciones que puede hacer el robot).

        7.2 El problema que se quiere resolver (estado inicial, estado objetivo y objetos del mundo representados).

        7.3 Introduzca el código desarrollado en los puntos anteriores en el [planificador online](http://lcas.lincoln.ac.uk/fast-downward/) y obtenga el plan de acción que tomará el robot para cumplir lo solicitado.


<h3> Resolución </h3>
  7.1 Dominio del Robot en Lenguaje STRIPS:

  Predicados:

    En(Oficina):              Indica que el robot está en la oficina.
    En(Tienda):               Indica que el robot está en la tienda.
    En(Buzón):                Indica que el robot está en el buzón.
    Tiene(Café):              Indica que el robot tiene una taza de café.
    Tiene(Carta):             Indica que el robot tiene una carta.
    Tiene(Caja):              Indica que el robot tiene una caja que puede contener objetos.
    ObjetoEn(Caja, Objeto):   Indica que un objeto está dentro de la caja.
    ObjetoEn(Oficina, Carta): Indica que la carta está en la oficina.
    ObjetoEn(Tienda, Café):   Indica que la taza de café está en la tienda.
    
  Acciones:

    Mover(Origen, Destino):   Permite al robot moverse de un lugar a otro.
    Tomar(Objeto):            Permite al robot tomar un objeto del lugar en el que se encuentra.
    Depositar(Objeto, Lugar): Permite al robot depositar un objeto en un lugar.
    ComprarCafé():            Permite al robot comprar una taza de café en la tienda.
    TomarCarta():             Permite al robot tomar una carta del buzón.

7.2 Problema a Resolver:

  Estado Inicial:

    El robot está en la oficina (En(Oficina)).
    El café está en la tienda (ObjetoEn(Tienda, Café)).
    La carta está en la oficina (ObjetoEn(Oficina, Carta)).
    El robot no tiene ningún objeto (¬Tiene(Café), ¬Tiene(Carta)).
    El robot no tiene una caja (¬Tiene(Caja)).
    No hay objetos en la caja (¬ObjetoEn(Caja, _)).
  
  Estado Objetivo:

    El robot está en la oficina (En(Oficina)).
    El robot tiene una taza de café (Tiene(Café)).
    El robot tiene una carta (Tiene(Carta)).
  
  Objetos del Mundo Representados:

    Oficina
    Tienda
    Buzón
    Café
    Carta
    Caja
    
  7.3 Codigo desarrollado

    (define (domain robot_office)
      (:predicates
        (En ?lugar)
        (Tiene ?objeto)
        (ObjetoEn ?lugar ?objeto)
        (TieneCaja ?robot)
      )

      (:action Mover
        :parameters (?origen ?destino)
        :precondition (and (En ?origen) (not (= ?origen ?destino)))
        :effect (and (En ?destino) (not (En ?origen)))
      )

      (:action Tomar
        :parameters (?objeto ?lugar ?robot)
        :precondition (and (En ?lugar) (ObjetoEn ?lugar ?objeto) (not (Tiene ?objeto)) (or (not (TieneCaja ?robot)) (and (TieneCaja ?robot) (Tiene ?objeto))))
        :effect (and (Tiene ?objeto) (not (ObjetoEn ?lugar ?objeto)))
      )

      (:action TomarCaja
        :parameters (?lugar ?robot)
        :precondition (and (En ?lugar) (ObjetoEn ?lugar Caja) (not (TieneCaja ?robot)))
        :effect (and (TieneCaja ?robot) (not (ObjetoEn ?lugar Caja)))
      )

      (:action ComprarCafé
        :precondition (and (En Tienda) (ObjetoEn Tienda Café))
        :effect (and (Tiene Café) (not (ObjetoEn Tienda Café)))
      )
    )

    (define (problem robot_problem)
      (:domain robot_office)
      (:objects Oficina Tienda Buzón Café Carta Caja Robot)
      (:init
        (En Oficina)
        (ObjetoEn Tienda Café)
        (ObjetoEn Buzón Carta)
        (ObjetoEn Oficina Caja)
        (not (Tiene Café))
        (not (Tiene Carta))
        (not (Tiene Caja))
        (not (TieneCaja Robot))
      )
      (:goal (and
        (En Oficina)
        (Tiene Café)
        (Tiene Carta)
      ))
    )
