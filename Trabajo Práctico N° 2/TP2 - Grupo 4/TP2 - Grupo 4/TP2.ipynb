<h2>Ejercicios Teóricos

1. ¿Qué diferencia hay entre una estrategia de búsqueda Informada y una estrategia de búsqueda No Informada?

  La Búsqueda informada (también conocida como búsqueda heurística) es un enfoque en la resolución de problemas y algoritmos de búsqueda que utiliza información adicional al objetivo, a menudo en forma de heurísticas y/o estimaciones, para guiar la exploración del espacio de estados en busca de soluciones, priorizando los estados que parecen más prometedores según la información proporcionada.

2. ¿Qué es una heurística y para qué sirve?

  Una **heurística** es una técnica o estrategia que proporciona informacion sobre una solución aproximada o subóptima a un problema complejo, cuando la solución exacta u óptima es difícil o costosa de encontrar. Las heurísticas se basan en reglas empíricas, intuiciones o conocimiento previo del problema y se utilizan para tomar decisiones informadas en situaciones donde la exploración exhaustiva de todas las posibilidades no es práctica.
  Por lo tanto, una heurística sirve para simplificar problemas difíciles y encontrar soluciones aceptables en un tiempo razonable.

3. Analice cuántos estados posibles puede adoptar el Mundo de la Aspiradora si existieran 3 habitaciones.

  Si tenemos 3 habitaciones, con dos estados posibles (sucio o limpio), entonces tenemos $2^3 = 8$ estados posibles:

  - Habitación 1: Sucia, Habitación 2: Sucia, Habitación 3: Sucia
  - Habitación 1: Sucia, Habitación 2: Sucia, Habitación 3: Limpia
  - Habitación 1: Sucia, Habitación 2: Limpia, Habitación 3: Sucia
  - Habitación 1: Sucia, Habitación 2: Limpia, Habitación 3: Limpia
  - Habitación 1: Limpia, Habitación 2: Sucia, Habitación 3: Sucia
  - Habitación 1: Limpia, Habitación 2: Sucia, Habitación 3: Limpia
  - Habitación 1: Limpia, Habitación 2: Limpia, Habitación 3: Sucia
  - Habitación 1: Limpia, Habitación 2: Limpia, Habitación 3: Limpia

  Luego, la aspiradora puede estar en cualquiera de las 3 habitaciones, por lo que tenemos $8 * 3 = 24$ estados posibles.

4. ¿Es posible que un algoritmo de búsqueda no tenga solución?
    
  Si, es posible que un algoritmo de búsqueda no tenga solución. Esto puede ocurrir cuando el algoritmo no puede encontrar un camino desde el estado inicial al estado objetivo y esto puede deberse a varias razones. En primer lugar, podría ser que no exista ningún camino válido que lleve desde el estado inicial hasta el estado objetivo en el espacio de estados. En segundo lugar, aunque un exista camino válido, podría ser infinitamente costoso en términos de recursos, tiempo o distancia, y el algoritmo de búsqueda podría ser incapaz de explorar ese camino dentro de un tiempo razonable.
  
  Es importante destacar que si un algoritmo de búsqueda no encuentra una solución, no necesariamente implica que el problema no tenga una soluciion valida.
