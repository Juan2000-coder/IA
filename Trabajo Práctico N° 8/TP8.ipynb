2. ¿Qué diferencia principal existe entre un algoritmo supervisado y un algoritmo no supervisado?

La diferencia principal entre un algoritmo supervisado y uno no supervisado es la etiquetación. Un algoritmo supervisado se entrena con datos que están etiquetados, es decir, se conoce la identidad de cada uno de los datos u objetos. En cambio un algoritmo no supervisado se entrena con datos no etiquetados, es decir, no se conoce la identidad de los datos. Por es motivo un algoritmo supervisado puede determinar la identidad de un nuevo dato mientras que uno no supervisado puede determinar solo la pertenencia de ese nuevo dato a un conjunto no identificado.

```python
#TP N°8 - Punto 4
import numpy as np

# Parámetros del Q-learning
gamma = 0.9          # Factor de descuento: Es el peso que se le da a las recompensas futuras.
#Si el factor de descuento es cercano a 1: El agente prefiere recompensas a largo plazo
#Si el factor de descuento es cercano a 0: El agente prefiere recompensas a corto plazo
#El objetivo del valor de gamma es encontrar un balance entre recompensas a corto y largo plazo
#Permite que el agente elimine la incertidumbre.

alpha = 0.1          # Tasa de aprendizaje
# La tasa de aprendisaje es un valor entre 0 y 1 que determina la rapidez con la que el agente aprende.
# Si la tasa de aprendizaje es cercana a 0: El agente aprende lentamente
# # Si la tasa de aprendizaje es cercana a 1: El agente aprende rápidamente


# Matriz de recompensa (6x6)
rewards = np.array([
    [0, 30, -1, 70, -1, -1],
    [30, 0, 70, -1, -1, -1],
    [-1, 0, 0, 70, -1, 70],
    [0, -1, 70, 0, 100, -1],
    [-1, -1, -1, 0, 100, 0],
    [-1, -1, 0, -1, 100, 0]
])

# Utilizando los valores de recompensa que utilizamos en clases.


# Q-Tabla inicial (6x6), todas las entradas a 0
q_table = np.zeros((6, 6)) # Inicializar Q-Tabla

# Algoritmo Q-learning
convergence_threshold = 0.01  # Umbral de convergencia: Es el valor máximo de cambio en la Q-Tabla para considerar que el algoritmo ha convergido
max_episodes = 10000      # Número máximo de episodios: Es el número máximo de iteraciones del algoritmo

old_q_table = np.copy(q_table)  # Inicializar old_q_table

for episode in range(max_episodes):
    state = np.random.randint(6)  # Elegir aleatoriamente un estado inicial
    while True:  # Loop hasta que alcancemos el estado objetivo
        possible_actions = np.where(rewards[state] >= 0)[0]  # Acciones permitidas: Si la recompensa es negativa, la acción no está permitida
        action = np.random.choice(possible_actions)  # Seleccionar una acción permitida aleatoriamente: Si hay más de una acción permitida, se elige una al azar
        next_state = action # El siguiente estado es el mismo que la acción elegida
        reward = rewards[state, action] # Recompensa de la acción elegida
        q_table[state, action] = q_table[state, action] + alpha * (reward + gamma * np.max(q_table[next_state, :]) - q_table[state, action])   # Actualizar Q-Tabla
        state = next_state # Actualizar estado

        if state == 5:  # Si hemos alcanzado el estado objetivo, terminamos el episodio
            break

    if episode % 100 == 0: # Verificar la convergencia cada 100 episodios, cada 100 pq se vuelve lento si es menos.

        max_change = np.max(np.abs(q_table - old_q_table)) # Calcular el cambio máximo en la Q-Tabla
        if max_change < convergence_threshold: # Si el cambio máximo es menor al umbral de convergencia, terminamos el algoritmo
            break
    old_q_table = np.copy(q_table)




# Normalizar la Q-Tabla
q_table_normalized = (q_table - np.min(q_table)) / (np.max(q_table) - np.min(q_table))

# Redondear la Q-Tabla a tres decimales
q_table_rounded = np.round(q_table_normalized, 3)

print(f"Convergencia alcanzada en {episode} episodios.")
print("Q-Tabla normalizada (redondeada a dos decimales):")
print(q_table_rounded)

```
