1. ¿Cuáles son los tres axiomas de Kolmogorov?

Los axiomas de Kolmogorov son:

1 - La probabilidad de un suceso es positiva.

2 - La probabilidad del suceso seguro es igual a 1

3 - La probabilidad de un conjunto de sucesos mutuamente excluyentes es igual a la suma de las probabilidades de todos ellos.
